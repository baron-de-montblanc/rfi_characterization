```python
##Importing data files
with open('Data/temp_dict.yaml', 'r') as file:
    temps = yaml.unsafe_load(file)

with open('Data/med_ssins.yaml', 'r') as file:
    meds = yaml.unsafe_load(file)

with open('Data/pointing.yaml', 'r') as file:
    points = yaml.safe_load(file)

#Making mask files for where the pointings are
p3 = np.where(np.array(list(points.values()))==3, True, False)
p1 = np.where(np.array(list(points.values()))==1, True, False)
p0 = np.where(np.array(list(points.values()))==0, True, False)
p2 = np.where(np.array(list(points.values()))==2, True, False)
p4 = np.where(np.array(list(points.values()))==4, True, False)


##Making the keys into strings
temps = {str(key): value for key, value in temps.items()}


##Constructing our filter keys
x = [i for s in clean_files_p3 for i in s]
matching_keys = [key for key in temps if any(s in key for s in x)]
print(matching_keys)
print(len(temps))


##Building the filtered temperature, pointing dictionaries
filt_temp = {key: temps[key] for key in temps if key in matching_keys}
filt_point = {key: points[key] for key in points if key in matching_keys}

##Labelling our data to isolate each pointing
tags, num = label(np.array(list(filt_point.values())))
print(tags.shape)


mean_temps = np.array([])
for i in range(num):
    mean_temps = np.append(mean_temps, np.mean(np.array(list(filt_temp.values()))[tags == i]))

print(mean_temps.shape)

#plt.scatter(np.array(list(temps.values()))[p3], np.array(list(meds.values()))[p3], s=1, label='3')
plt.scatter(np.array(list(temps.values()))[p1], np.array(list(meds.values()))[p1], s=1, label='1')
#plt.scatter(np.array(list(temps.values()))[p0], np.array(list(meds.values()))[p0], s=1, label='0')
#plt.scatter(np.array(list(temps.values()))[p2], np.array(list(meds.values()))[p2], s=1, label='2')
plt.scatter(np.array(list(temps.values()))[p4], np.array(list(meds.values()))[p4], s=1, label='4')
plt.legend()
plt.show()

plt.scatter(np.arange(0, mean_temps.shape[0]), mean_temps)
plt.show()

fig, ax = plt.subplots(figsize=(10, 10), ncols=1, nrows=5, dpi=600)
ax[0].hist(np.array(list(meds.values()))[p3], bins='auto', density=True, histtype='step')
ax[1].hist(np.array(list(meds.values()))[p1], bins='auto', density=True, histtype='step')
ax[2].hist(np.array(list(meds.values()))[p0], bins='auto', density=True, histtype='step')
ax[3].hist(np.array(list(meds.values()))[p2], bins='auto', density=True, histtype='step')
ax[4].hist(np.array(list(meds.values()))[p4], bins='auto', density=True, histtype='step')
ax[0].set_xticks([])
ax[1].set_xticks([])
ax[2].set_xticks([])
ax[3].set_xticks([])
fig.set_tight_layout(tight=True)
fig.supylabel('Density')
fig.supxlabel('Amplitude (arb. units)')
plt.show()
```
