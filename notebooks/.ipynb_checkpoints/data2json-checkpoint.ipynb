## Send MWA data to JSON format for training

```python
import numpy as np
import json
import h5py
import matplotlib.pyplot as plt
```

```python
mwa_path = '/users/jmduchar/data/shared/mwa_fall_16/dataset/all/'
file = '1160763448_4165wcu_Pxx.hdf5'

with h5py.File(mwa_path+file, "r") as hf:
    capture_group = hf["capture"]

    # Metadata
    obsid = file.split('_')[0]
    pol = file.split('_')[-1].split(".")[0][1:]
    print("OBSID:", obsid)
    print("Polarization:", pol)
    print("All HDF5 Keys:", capture_group.keys())

    # Access the SSINS
    ssins = capture_group['ssins_rep'][:]
```

```python
plt.figure(figsize=(6,3))
plt.imshow(ssins, aspect='auto', cmap='viridis')
plt.colorbar()
plt.show()
```

```python
# AVERAGE ACROSS DTV-7
ssins_avg = np.mean(ssins[:,350:525], axis=1)

ssins_avg.shape
```

```python
plt.plot(ssins_avg)
```

```python
# Structure the data for Stan
stan_data = {
    "T": ssins_avg.shape[0],
    "y": ssins_avg.tolist(),
    "mu0_clean": 0,  # prior on clean mean
    "sigma0_clean": 1,  # prior on clean std
}
```

```python
stan_data
```

```python
# Write to JSON
with open("two_state.data.json", "w") as f:
    json.dump(stan_data, f)
```
